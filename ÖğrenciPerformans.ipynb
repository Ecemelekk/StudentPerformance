```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Veriyi oku
df = pd.read_csv('/content/student_data.csv') # Dosya adının tam olduğundan emin ol

# İlk 5 satıra bak
print(df.head())

# İstatistiksel özet (Sayısal sütunların ortalaması, minimumu, maksimumu)
print(df.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Görselleştirme stilini ayarla (Profesyonel bir görünüm için)
sns.set_theme(style="whitegrid")

# Analiz: Aile desteği alan vs almayan öğrencilerin Final Notu karşılaştırması
plt.figure(figsize=(10, 6))

# Boxplot ile başarı dağılımını ve medyan farkını gösteriyoruz
sns.boxplot(x='famsup', y='G3', data=df, palette='pastel')

plt.title('Aile Desteğinin Öğrenci Başarısı Üzerindeki Etkisi', fontsize=15)
plt.xlabel('Aile Desteği Durumu', fontsize=12)
plt.ylabel('Final Notu', fontsize=12)

# Grafiği göster
plt.show()

# İstatistiksel özet: Aile desteği alanların ortalama notu nedir?
print("Aile Desteği Durumuna Göre Final Notu Ortalamaları:")
print(df.groupby('famsup')['G3'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Analiz: Yükseköğretim isteği ile Final Notu ilişkisi
plt.figure(figsize=(8, 5))
sns.boxplot(x='higher', y='G3', data=df, palette='Set2')
plt.title('Üniversiteye Gitme İsteği ve Final Notu İlişkisi')
plt.xlabel('Yükseköğretim İsteği')
plt.ylabel('Final Notu')
plt.show()

# 2. Analiz: Geçmiş başarısızlık sayısı ile Final Notu ilişkisi
plt.figure(figsize=(8, 5))
sns.boxplot(x='failures', y='G3', data=df, palette='magma')
plt.title('Geçmiş Başarısızlık Sayısının Final Notuna Etkisi')
plt.xlabel('Ders Başarısızlığı Sayısı')
plt.ylabel('Final Notu')
plt.show()

# Sayısal değerleri görelim
print("Üniversite İsteğine Göre Not Ortalamaları:")
print(df.groupby('higher')['G3'].mean())

print("\nGeçmiş Başarısızlık Sayısına Göre Not Ortalamaları:")
print(df.groupby('failures')['G3'].mean())
```
